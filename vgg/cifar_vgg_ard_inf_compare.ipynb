```python
import torch

filelist=[
    'ckpt/cifar10_vggD6_ard_13-history.ckpt', 
    'ckpt/cifar10_vggG5_ard_13-history.ckpt',
    'ckpt/cifar100_vggG_ard_13-history.ckpt',
    'ckpt/cifar100_vggG5_ard_13-history.ckpt'
]

legendlist=[
    'VGG-Torch on CIFAR-10',    
    'VGG-G5 on CIFAR-10',    
    'VGG-Torch on CIFAR-100',    
    'VGG-G5 on CIFAR-100',
    
]
epochn=[8,5,7,6]

for idd in range(len(filelist)):
    ckptfile=filelist[idd]
    history=torch.load(ckptfile)
    errlist=history[0]
    edgeratiolist=history[1]
    covlist=history[2]
    linearlist=history[3]
    
    topk=epochn[idd]
    
    print(' \multirow{',end=' ') 
    print(topk,end=' ')
    print('}{2.3cm}{',end=' ')
    print(legendlist[idd],end=' ')
    print('}')
    
    
    for epoch in range( topk):
        err=errlist[epoch]
        leftcovnode=covlist[epoch]
        leftlinearnode=linearlist[epoch]
        edgeratio=edgeratiolist[epoch]
        print('& @',end='')
        print(epoch+1,end='\t')
        print('&',end=' ')
        for n in range(len(leftcovnode)-1):
            print(leftcovnode[n],end='-')
        print(leftcovnode[-1],end=' ')
        for n in range(len(leftlinearnode)-1):
            print(leftlinearnode[n],end='-')
        print(leftlinearnode[-1],end='\t')
        print('& {:.2f} &\t{:.2f} \\\\ '.format(edgeratio,err))
    print('\hline')
    print()
```
